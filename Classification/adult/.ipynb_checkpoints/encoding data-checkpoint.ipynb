```python
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter('ignore')
```

```python
data = pd.read_csv('trimmed_census_data.csv')
data.drop(columns='Unnamed: 0',axis=1,inplace=True)
data.head()
```

```python
data.info()
```

```python
data_encoded = pd.get_dummies(data, columns=['workclass','marital-status','relationship','race','sex'],drop_first=True)
```

```python
data_encoded.head()
```

```python
# education and edu-num is same

data_encoded.drop('education',axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
data_encoded['occupation'] = LabelEncoder().fit_transform(data_encoded.occupation)
data_encoded['native-country'] = LabelEncoder().fit_transform(data_encoded['native-country'])
data_encoded.drop(['occupation', 'native-country'], axis=1, inplace=True)
data_encoded.head()
```

### exporting csv

```python
data_encoded.to_csv('census_data_encoded.csv')
```
